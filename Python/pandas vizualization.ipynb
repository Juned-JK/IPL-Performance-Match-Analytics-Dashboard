```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\juned\Videos\data analytics\end to end projects\IPL DATASET\Data\IPL.csv",low_memory=False )
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
top10batsman = df.groupby('batter')['runs_batter'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top10batsman, y='batter', x='runs_batter')
plt.title('Top 10 Batsman')
plt.show()
```

```python
top10bowlers = df.groupby('bowler')['bowler_wicket'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top10bowlers, y='bowler', x='bowler_wicket')
plt.title('Top 10 bowlers')
plt.show()
```

```python
runsbyseason = df.groupby('season')['runs_batter'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=runsbyseason , y = 'season', x = 'runs_batter')
plt.title('Total runs by Season')
plt.show()
```

```python
wicketsbyseason = df.groupby('season')['bowler_wicket'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=wicketsbyseason, y = 'season', x = 'bowler_wicket')
plt.title('Total wickets by season')
plt.show()
```

```python
team_sixes = df[df['runs_batter'] == 6].groupby('batting_team').size().sort_values(ascending=False).reset_index(name='total_sixes')
plt.figure(figsize=(10,6))
sns.barplot(data=team_sixes, y='batting_team', x='total_sixes')
plt.title('Total sixes by each team')
plt.show()
```

```python
team_sixes = df[df['runs_batter'] == 4].groupby('batting_team').size().sort_values(ascending=False).reset_index(name='total_fours')
plt.figure(figsize=(10,6))
sns.barplot(data=team_sixes, y='batting_team', x='total_fours')
plt.title('Total fours by each team')
plt.show()
```

```python
matchesbyvanue = df.groupby('venue')['match_id'].nunique().sort_values(ascending = False).reset_index()
plt.figure(figsize=(10,10))
sns.barplot(data=matchesbyvanue, y='venue', x='match_id')
plt.title('Total number of matches by venue')
plt.show()
```

```python
strike_rate = (df.groupby('batter').agg(total_runs=('runs_batter', 'sum'),balls_faced=('valid_ball', 'sum')).reset_index())
strike_rate['strike_rate'] = (strike_rate['total_runs'] / strike_rate['balls_faced']) * 100
strike_rate = strike_rate.sort_values(by='strike_rate', ascending=False)
strike_rate.head()
```

```python
top_sr = (strike_rate[strike_rate['balls_faced'] >= 100].sort_values(by='strike_rate', ascending=False).head(10))
plt.figure(figsize=(10, 6))
sns.barplot(data=top_sr, x='strike_rate', y='batter')
plt.title('Top 10 Batters by Strike Rate (Min. 100 Balls Faced)')
plt.xlabel('Strike Rate')
plt.ylabel('Batter')

plt.show()
```

```python
economy_rate = df.groupby('bowler').agg(total_runs=('runs_batter', 'sum'),total_balls=('valid_ball', 'sum')).reset_index()
economy_rate['economy'] = (economy_rate['total_runs'] * 6 / economy_rate['total_balls'])
economy_rate = economy_rate.sort_values(by = 'economy')
economy_rate.head(10)
```

```python
top_economy = (economy_rate[economy_rate['total_balls'] >= 24].sort_values(by='economy', ascending=True).head(10))
plt.figure(figsize=(10, 6))
sns.barplot(data=top_economy, x='economy', y='bowler')
plt.title('Top 10 Bowlers by Economy Rate (Min. 24 Balls bowled)')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')

plt.show()
```

```python
orange_cap = (df.groupby(['season', 'batter'])['runs_batter'].sum().reset_index(name='total_runs'))
orange_cap = ( orange_cap.sort_values(['season', 'total_runs'], ascending=[True, False]).groupby('season').head(1))

plt.figure(figsize=(12,6))
sns.lineplot(data=orange_cap,x='season',y='total_runs',marker='o')
plt.title('Orange Cap Trend by Season')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.xticks(rotation=45)
plt.show()
```

```python
purple_cap = (df.groupby(['season', 'bowler'])['bowler_wicket'].sum().reset_index(name='total_wickets'))
purple_cap = (purple_cap.sort_values(['season', 'total_wickets'], ascending=[True, False]).groupby('season').head(1))

plt.figure(figsize=(12,6))
sns.lineplot(data=purple_cap,x='season',y='total_wickets',marker='o')
plt.title('Purple Cap Trend by Season')
plt.xlabel('Season')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)
plt.show()
```
